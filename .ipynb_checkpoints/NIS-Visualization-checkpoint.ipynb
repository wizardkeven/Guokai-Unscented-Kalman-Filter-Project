```python
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
```

## Read in NIS data and visulize ##

```python
import csv

nis_of_lidar = []
with open('build/NIS_OF_LIDAR.csv', newline = '') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for row in reader:
        nis_of_lidar.append(row[0])

plt.axhline(y=5.991, color='r', linestyle='-')
plt.plot(nis_of_lidar)
plt.title("NIS graph of Lidar")
plt.show()
```

```python
nis_of_radar = []
with open('build/NIS_OF_RADAR.csv', newline = '') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for row in reader:
        nis_of_radar.append(row[0])

plt.axhline(y=7.815, color='r', linestyle='-')
plt.plot(nis_of_radar)
plt.title("NIS graph of Radar")
plt.show()
```
